```python
from math import floor as fl
import requests
from bs4 import BeautifulSoup
import pandas as pd


"""Get the article from the url given"""
def get_article(url):
    text  = []
    r = requests.get(url)

    soup = BeautifulSoup(r.content, "html.parser")
    main = soup.find("main", {"class": "site-main"}).find("article")
    for i in main.find_all("p"):
        text.append(i.text)
    return ' '.join(text)

"""Get the total number of pages"""
def get_total_pages(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    total_result = soup.find("main", {"class": "site-main"}).find("p").find("strong").text
    if total_result == int(0):
        print ("article not found")
    total_pages = fl(int(total_result)/12)
    return total_pages

def get_articles(searchterms:str):
    if searchterms == "":
        print("Please enter a search term")
        get_articles(input("Enter search term: "))
    if " " in searchterms:
        searchterms = searchterms.replace(" ", "+")
    url = f"https://www.housingwire.com/?s={searchterms}"
    total_pages = get_total_pages(url)
    print(f"there are {total_pages} pages to complete")
    review_list = []
    for pages in range(1,total_pages+1):
        print(f"{pages} of {total_pages}")
        page_url = f"https://www.housingwire.com/page/{pages}/?s={searchterms}"
        r = requests.get(page_url)
        soup = BeautifulSoup(r.content, "html.parser")
        main = soup.find("main", {"class": "site-main"})
        for i in main.find_all("article"):
            title = i.find("a",{"class":"entry-title"}).text
            article_link = i.find("a",{"class":"entry-title"})['href']
            if "podcast".lower() not in article_link.lower():
                article = get_article(article_link)
                date = i.find("span",{"class":"date-posted"}).text
                author = i.find("span",{"class":"author"}).find("a").text
                review_list.append([date,author,title,article,article_link])
    return review_list

articles = get_articles("home depot")

df = pd.DataFrame(articles, columns=['date', 'author','title','article','article_link'])
```
